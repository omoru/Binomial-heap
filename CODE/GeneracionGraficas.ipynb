```python

#Cargar tiempos
with open('/content/times.txt') as f:
    tiempos = [ float(x) for x in f]
```

```python
#Lo tenemos en nanosegundos, transformamos a milisegundos
tiempos = [t/1000000 for t in tiempos]
```

```python
#Cargar tama√±os
with open('/content/tamanios.txt') as f:
    tamanios = [int(x) for x in f]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(40, 20))
plt.ylim(0, 2* max(tiempos))
#plt.title("DecreaseKey", size = 26)
plt.ylabel("Time (ms)", size = 26,)
plt.xlabel("Number of nodes", size = 26)
plt.plot(tamanios,tiempos)
```
